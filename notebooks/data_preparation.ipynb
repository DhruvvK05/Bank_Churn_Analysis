## Feature Engineering 

Creating customer age segments and other Columns to Make visualization better and get better insights to analyze churn behavior across different age demographics.

```python
df['age_group'] = pd.cut(
    df['age'],
    bins=[18, 30, 40, 50, 60, 100],
    labels=['18-30', '31-40', '41-50', '51-60', '60+']
)
```

```python
df[['age', 'age_group']].head()
```

```python
df['salary_band'] = pd.qcut(
    df['estimated_salary'],
    q=4,
    labels=['Low', 'Medium', 'High', 'Very High']
)
```

```python
df[['estimated_salary', 'salary_band']].head()
```

```python
df['balance_group'] = pd.cut(
    df['balance'],
    bins=[-1, 50000, 100000, 150000, 300000],
    labels=['Low', 'Medium', 'High', 'Very High']
)
```

```python
df[['balance', 'balance_group']].head()
```

```python
df['churn_status'] = df['exited'].map({
    0: 'Retained',
    1: 'Churned'
})
```

```python
df['activity_status'] = df['is_active_member'].map({
    0: 'Inactive',
    1: 'Active'
})
```
